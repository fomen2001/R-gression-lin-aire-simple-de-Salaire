```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression









```

```python
# 1. Importation et chargement [cite: 97, 98]
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:, :-1].values  # Années d'expérience
y = dataset.iloc[:, -1].values   # Salaire
```

```python
dataset.head()  # Affiche les premières lignes du dataset
```

```python
dataset.describe()  # Affiche les statistiques descriptives du dataset
```

```python
# 3. Division du jeu de données (2/3 train, 1/3 test) [cite: 100, 101]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)
```

```python
# 4. Construction et entraînement du modèle [cite: 103, 104]
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# 5. Prédictions [cite: 105]
y_pred = regressor.predict(X_test)
# Prédiction pour 15 ans d'expérience
prediction_15 = regressor.predict([[15]])
print(f"Salaire prédit pour 15 ans d'expérience : {prediction_15[0]}")
```

```python
# 6. Visualisation [cite: 106, 107, 108]
plt.scatter(X_test, y_test, color='red') # Données réelles
plt.plot(X_train, regressor.predict(X_train), color='blue') # Ligne de régression
plt.title('Salaire vs Expérience (Ensemble de Test)')
plt.xlabel('Années d expérience')
plt.ylabel('Salaire')
plt.show()
```

3. Réponses aux Questions de Discussion

Signification de la régression linéaire simple C'est un outil statistique qui permet de modéliser la relation entre deux variables par une ligne droite. Elle permet de comprendre comment la valeur de la variable dépendante change lorsque la variable indépendante varie.

Entraînement du modèle Le modèle est entraîné en utilisant les données de l'ensemble d'apprentissage. Il calcule mathématiquement les coefficients de la droite de manière à minimiser la somme des carrés des distances entre les points réels et la droite.

Ensemble d'apprentissage vs Ensemble de test Apprentissage : Sert à "éduquer" le modèle pour qu'il trouve la corrélation.Test : Sert à évaluer la performance du modèle sur des données qu'il n'a jamais vues.Importance : Cette division permet de vérifier si le modèle est capable de généraliser ses prédictions et d'éviter qu'il ne fasse que "mémoriser" les données d'entraînement (surapprentissage).

```python
import joblib

# Sauvegarde du modèle entraîné dans un fichier nommé 'model_salaire.pkl'
joblib.dump(regressor, 'model_salaire.pkl')

print("Modèle sauvegardé avec succès !")
```
